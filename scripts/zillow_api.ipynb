```python
import requests
import pandas as pd
```

## Define functions

```python
def get_property_info(location):
    url = "https://zillow56.p.rapidapi.com/search"
    
    querystring = {"location": location}

    headers = {
        "X-RapidAPI-Key": "9fb3ebb517mshef1c2cc7b5f6c16p16e0e4jsn72dbf0c17917",
        "X-RapidAPI-Host": "zillow56.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)

    if response.status_code == 200:
        data = response.json()

        results = data.get('results', [])

        for property_info in results:
            bathrooms = property_info.get('bathrooms')
            bedrooms = property_info.get('bedrooms')
            city = property_info.get('city')
            price = property_info.get('price')
            street_address = property_info.get('streetAddress')

            print(f"Property: {street_address}, {city}")
            print(f"Bedrooms: {bedrooms}, Bathrooms: {bathrooms}")
            print(f"Price: ${price}")
            print("="*30)
    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")


```

```python
def get_property_details(zpid):
    url = "https://zillow56.p.rapidapi.com/property"
    
    querystring = {"zpid": zpid}

    headers = {
        "X-RapidAPI-Key": "9fb3ebb517mshef1c2cc7b5f6c16p16e0e4jsn72dbf0c17917",
        "X-RapidAPI-Host": "zillow56.p.rapidapi.com"
    }

    response = requests.get(url, headers=headers, params=querystring)

    if response.status_code == 200:
        property_details = response.json()
        if property_details:
            property_df = pd.json_normalize(property_details)
            return property_df
        else:
            print("No data retrieved.")
#         return property_details
    else:
        print(f"Failed to retrieve data. Status code: {response.status_code}")
        return None
```

```python
def flatten_json(json_obj, prefix=''):
    flattened_dict = {}
    for key, value in json_obj.items():
        new_key = f"{prefix}_{key}" if prefix else key
        if isinstance(value, dict):
            flattened_dict.update(flatten_json(value, new_key))
        else:
            flattened_dict[new_key] = value
    return flattened_dict
```
